# Variance Inflation Factor(VIF):
    Used for finding multi colenearity.So,that we can remove it

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
def cal_vif(ds):
    vif=pd.DataFrame()
    vif["variables"]=ds.columns
    vif["VIF FACTOR"]=[variance_inflation_factor (ds.values,i) for i in range (ds.shape[1])]
    return (vif)

cal_vif(df)
```

```python
df.drop("veil-color",axis=1,inplace=True)
```

# Scaling

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
```

Import all the nessary lib needed

```python
for i in range(200):
    newX_train,newX_test,newY_train,newY_test=train_test_split(newX,newY,test_size=.25,random_state=i)
    dtc=DecisionTreeClassifier()
    dtc.fit(newX_train,newY_train)
    pred=dtc.predict(newX_test)
    acc=accuracy_score(newY_test,pred)
    print(acc*100,'at_random_state',i)
```

```python
maxAccu=0
maxRS=0
for i in range (1,200):
    newX_train,newX_test,newY_train,newY_test= train_test_split(newX,newY,test_size=.30,random_state =i)
    dtc.fit(newX_train,newY_train)
    pred = dtc.predict(newX_test)
    acc=accuracy_score(newY_test,pred)
    if acc>maxAccu:
        maxAccu=acc
        maxRS=i

print("Best accuracy is ", maxAccu," on Random state" ,maxRS)
```
